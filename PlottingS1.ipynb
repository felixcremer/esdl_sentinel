```julia
dB(x) = 10 * log10(x)
```

```julia
asc_dB = map(dB, data_asc)
```

```julia
trend_db = loadCube("generation/cubes/s1cube_asc_trend_dB/")
```

```julia
LONORIGIN =-98.6871269689448098
LATORIGIN = 20.6917955161275380
resolution = 0.000179663056824 
```

```julia
function copyfun(xout, xin, inds)
    xout .= xin[inds]
end
```

```julia
LatAxis(trend_dB)
```

```julia
latshift = LATORIGIN - trend_dB.axes[2].values[1]
trend_dB.axes[2].values .+ latshift
```

```julia
newlataxis = LatAxis(trend_dB.axes[2].values .+ latshift)
```

```julia
# prepare mapCube
indims = InDims("Lat")
outdims = OutDims(newlataxis)

# rewrite the axis
newlatcube = mapCube(copyfun, trend_dB, collect(1:length(newlataxis)), indims=indims, outdims=outdims, max_cache=1e8)
```

```julia
lonshift = LONORIGIN - trend_dB.axes[2].values[1]
trend_dB.axes[2].values .+ lonshift
```

```julia
newlonaxis =  LonAxis(trend_dB.axes[2].values .+ lonshift)
```

```julia
indims = InDims("Lon")
outdims = OutDims(newlonaxis)
newloncube = mapCube(copyfun, newlatcube, collect(1:length(newlonaxis)), indims=indims, outdims=outdims, max_cache=1e8)
```

```julia
exportcube(newloncube, "trend_georef_2.nc")
```

```julia
cube = Cube("/media/data/esdl/s1cube_1dayallnew/")
```

```julia
Ax
```

```julia
latshift = LATORIGIN - data_asc.latAxis.values[1] - 0.5*resolution
newlataxis = LatAxis(data_asc.latAxis.values .+ latshift)
```

```julia
lonshift = LONORIGIN - data_asc.lonAxis.values[1]
newlonaxis = LonAxis(data_asc.lonAxis.values .+ lonshift)
```

```julia
# prepare mapCube
indims = InDims("Lat")
outdims = OutDims(newlataxis)
# rewrite the axis
newlatcube = mapCube(copyfun, data_asc, collect(1:length(newlataxis)), indims=indims, outdims=outdims, max_cache=1e8)
```

```julia
indims = InDims("Lon")
outdims = OutDims(newlonaxis)
newloncube = mapCube(copyfun, newlatcube, collect(1:length(newlonaxis)), indims=indims, outdims=outdims, max_cache=1e8)
```

```julia
exportcube(newloncube, "asc_data_georef.nc")
```
